Importing csv files of latest 2 years of dataset

```python
c17 = pd.read_csv('/kaggle/input/pre-cog/cases/cases_2017.csv')
c18 = pd.read_csv('/kaggle/input/pre-cog/cases/cases_2018.csv')
```

Dropping irrelevant columns based on previous analysis

```python
c17.drop(['year', 'cino', 'date_of_filing', 'date_of_decision', 'date_first_list', 'date_last_list', 'date_next_list'], axis=1, inplace=True)
c18.drop(['year', 'cino', 'date_of_filing', 'date_of_decision', 'date_first_list', 'date_last_list', 'date_next_list'], axis=1, inplace=True)
```

Concatenating datasets

```python
cfinal = pd.concat([c17, c18], axis=0)
```

Exporting dataframe to a csv file

```python
cfinal.to_csv('case_year_mix.csv')
```
