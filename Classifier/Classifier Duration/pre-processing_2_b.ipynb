Importing the concatenated file

```python
cfinal = pd.read_csv('/kaggle/input/duration-set/year_mix_duration.csv')
print(cfinal.head())
```

Dropping the unnecessary columns (gender of petitioner and defendant have a lot of unclear, NaN, and unknown values which reduce the size of dataset heavily upon cleaning. Also the correlation found after using them for training to the target variable is very less. Hence they are being dropped)

```python
cfinal.drop(['Unnamed: 0', 'female_defendant', 'female_petitioner'], axis=1, inplace=True)
```

Importing the **Act and Sections** dataset

```python
act_sec = pd.read_csv('/kaggle/input/pre-cog/acts_sections/acts_sections.csv')
print(act_sec.head())
```

Dropping the column due to major data insufficiency

```python
act_sec.drop(['bailable_ipc'], axis=1, inplace=True)
```

Getting rid of the NaN values

```python
act_new = act_sec.dropna()
```

```python
act_new.reset_index(drop=True)
```

```python
print(cfinal.head())
```

Gender of the advocate does not influence case closing duration based on previous analysis

```python
cfinal.drop(['female_adv_def', 'female_adv_pet'], axis=1, inplace=True)
```

Null and empty value cleaning for the dataset

```python
cfinal['date_of_filing'].replace('', np.nan, inplace=True)
cfinal['date_of_decision'].replace('', np.nan, inplace=True)
```

Changing Data Type for the required columns

```python
cfinal["date_of_filing"] = pd.to_datetime(cfinal["date_of_filing"], errors = "coerce")
cfinal["date_of_decision"] = pd.to_datetime(cfinal["date_of_decision"], errors = "coerce")
```

```python
cfinal.dropna(subset=['date_of_filing'], inplace=True)
cfinal.dropna(subset=['date_of_decision'], inplace=True)
```

```python
cfinal.dtypes
```

Calculating the case closing duration for the filed cases

```python
cfinal['duration'] = cfinal['date_of_decision'] - cfinal['date_of_filing']
```

Merging the datasets based on the 'ddl_case_id' index

```python
data = cfinal.merge(act_new, left_on='ddl_case_id', right_on='ddl_case_id', how='left')
print(data.head())
```

```python
data_f = data.dropna()
```

```python
data_f.reset_index(drop=True)
```

```python
data_f.dtypes
```

Converting to number of days

```python
data_f['duration_f'] = data_f['duration'].dt.days
```

```python
data_f.dtypes
```

```python
data_f.drop(['duration'], axis=1, inplace=True)
```

```python
print(data_f.head())
```

Removing Outliers and inexplicable values

```python
data_f = data_f[(0 <= data_f['duration_f'])]
```

Exporting the final dataframe to a csv file

```python
data_f.to_csv('final_model_exp.csv')
```
